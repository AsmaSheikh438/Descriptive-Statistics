## Covariance and Correlation
* Both shows the realationship between Two Columns , In data science we use covariance and correlation to find out realtion between two columns with the help of this we can make decision which Algorithm will be best.

* The main difference between covariance and Correlation is RANGE.

## Positive and Negative Covariance

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('tip.csv')
dataset.head(4)
```

```python
# checking null values
dataset.isnull().sum()
```

```python
# no null values is present
```

```python
# checking data types first
dataset.info()
```

#### Finding Correlation

#### Correlation is used to find the relation between two columns. We use correlation in machine learning when we work on SIMPLE LINEAR REGRESSION .

```python
data_corr = dataset.select_dtypes(['float64','int64']).corr()
```

```python
# Plotting correlation graph

plt.figure(figsize=(4,3))
sns.heatmap(data_corr,annot= True)
plt.show()
```

#### Finding Covariance

```python
data_cov = dataset.select_dtypes(['float64','int64']).corr()
```

```python
dataset.select_dtypes(['float64','int64']).cov()
```

```python
# Plotting Covariance graph

plt.figure(figsize=(4,3))
sns.heatmap(data_cov,annot= True)
plt.show()
```
