## Random Variables
A Random Variable 'X' is a function that assigns a real number to each outcome in the sample space of a random experiment.

Random Variables are of two types..
### Discrete Random Variable :
A number variable that takes on a countable number of distinct values.
### Continuous Random Variable :
A Random Variable that can take on any value with a given range or interval.

# Probability Distribution

### Probability Distribution 
Probability Distributions describe how the probabilities of different outcomes are distributed over the sample space of a random variable.

Probability Distribution are of two types:
#### 1.  Discrete Probability Distribution
#### 2. Continuous Probability Distribution

## Probability Distribution Functions
A Probability Distribution is the mathematical function that gives the probabilities of occurance of different possible outcomes for an experiment.

###  Types of Probability Distribution Functions
#### 1. (PDF) probability Density Function
(PDF) use if data is CONTINUOUS
#### 2. (PMF) Probability Mass 
(PMF) use if data is DISCRETE
#### 3. (CDF) Cumalative Density Function
It gives information about the data present behind the particular range and data above the particular range, by adding the values simultaneously.

##### (Q). Why we move from Frequency Distribution chart to Probability Distribution Chart?

* If we only want to find out how many times a data occur in a particular range/Data then we use Frequency Distribution,

* But If we want to find out the probability/ Changes of data will be occured in next time in out in a particular range then, we use Probability Distribution Chart.
