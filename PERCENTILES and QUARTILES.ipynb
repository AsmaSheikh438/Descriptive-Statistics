# Percentiles
Percentiles are used in statistics to give a number to describes the value the the given percent of the values are lower than.

#### If we arrange the group of data in Ascending order, so what is the POSITION or RANKING of the group of data we calculate in percentiles. So we can say that the percentiles is the ranking system.
#### and we calculate this percentiles ranking system when we don't have any base.
#### Percentiles and Quartiles are used to find the Outliers when we work on datasets in Machine Learning.

## Quartiles
Quantiles divide the dataset into equal parts based on rank or percentile. They represent the values at certain points in a dataset sorted in increasing order. General quantiles include the median (50th percentile), quartiles (25th, 50th, and 75th percentiles), and percentiles (values ranging from 0 to 100).

### The given chart showing below in (IQR) called as BOX PLOT. It works on the percentiles bases.

# Interquartile Range (IQR)

### Outliers
Data beyond the range is called outliers.


#### When we get outliers in our data so, to find these outliers we use (IQR) that is Interquartile Range.
#### IQR is a method to find out Qutliers in our data.
#### IQR method work on percentiles

```python
# now working on real time data
# first importing libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('train.csv')
dataset.head(4)
```

```python
dataset.isnull().sum()
```

```python
dataset['Age'].fillna(dataset['Age'].mean(),inplace= True)
```

```python
dataset.isnull().sum()
```

```python
# now there is no null values in the age column.
```

```python
# Now we have to find the Percentiles
dataset['Age']
```

#### 0 percentile indicates the minimum value
#### 25 percentile indicates the First Quartile value (Q1)
#### 50 percentile indicate the Median value
#### 75 percentile indicates the Third Quartile (Q3)
#### 100 percentiles indicates the maximum value

##### 0 percentile = min value

```python
a = np.percentile(dataset['Age'],0)
a

b = dataset['Age'].min() # 0 percentile and minimum value both are same
b
```

##### 25 percentile is Q1

```python
c = np.percentile(dataset['Age'],25) # 25 percentile is Q1
c
```

##### 50 Percentile = median value

```python
d = np.median(dataset['Age'])
d
```

##### 75 percentile is Q3

```python
e = np.percentile(dataset['Age'],75)
e 
```

#### 100 percentile = max value

```python
f = np.percentile(dataset['Age'],100)
f

g = dataset['Age'].max()
g
```
